###### Section B.   Synopsize of assumptions for a time series model including stationarity and autocorrelated data. <a class="tocSkip">

###### B.1. Assumptions of time series models 

A primary assumption of time series modeling is 'Stationarity'.

In research from Mohr (2020), fundamentally stationarity explains that time series have a constant mean 
and a uniform variance over time. While not remarkably significant for the approximation of parameters 
of econometric models, these characteristics are necessary for the computation of trustworthy test
statistics, therefore, stationarity could have a substantial effect on model selection.

A seconary but equaly important time series assumption is 'Autocorrelated data'.

The autocorrelated assumption according to research from Hamilton (1994), is the same as when one generates a linear regression model in that the errors are independent. Specifically, the dependency 
typically materializes due to a temporal component. Error terms correlated over time are autocorrelated, or 
serially correlated.  If error terms are autocorrelated, issues may occur when using ordinary least squares.
Problems may be that estimated regression coefficients are still unbiased, but no longer maintain the minimum variance property required for accuracy.  Also, the MSE (mean squared error) may seriously 
undervalue the actual variance of the errors.  And the standard error of the regression coefficients can 
negatively underestimate the true standard deviation of the estimated regression coefficients.

According to new research (Chedalavada, 2021), non-stationary data is unpredictable in nature, by definition, and cannot be predicted or modelled.  If we employ this non-stationary time series, the results may be inauthentic, and may parody a relationship between variables that does not exist.

In research from Swanson (2016) there are four highly cited time series models used for forecasting that might provide a solution to our non-stationarity hindrance.  They are.

•	ARIMA model
•	Holt Winter’s Exponential Smoothing

ARMA model, according to new research from (Mohr, 2020), ARMA (Autoregressive Moving Average) model forecasts apply advanced econometric modeling techniques to forecast time-series data by first backfitting to historical data and then forecasting the future.


ARIMA model, Auto-regressive-integrated-moving-average (ARIMA) models rank as the fundamental pillar for time-series modeling.

Identified autoregressive integrated moving average (ARIMA) model, which counterbalances for observed trends and seasonality.

```python
model = sm.tsa.SARIMAX(data_train,
                       order=(3, 0, 2),
                       seasonal_order=(2, 1, 0, 12),
                       enforce_stationarity=False,
                       enforce_invertibility=False)

SARIMAX_results = model.fit()

print(SARIMAX_results.summary())
```

Forecast results

```python
result = SARIMAX_results.get_forecast()
# Summarize forecast and confidence intervals
test_1 = data_test['data_test'].values.astype('float32')
forecast = result.predicted_mean
print('Expected: %.2f' % forecast)
print('Forecast: %.2f' % test_1[0])
print('Standard Error: %.2f' % result.se_mean)
```

Examine Confidence Intervals

```python
intervals = [0.2, 0.1, 0.05, 0.01]
for a in intervals:
    ci = result.conf_int(alpha=a)
    print('%.1f%% Confidence Interval: %.2f between %.2f and %.2f' % ((1 - a) * 100, forecast, ci['lower data_train'], ci['upper data_train']))

ci
```
